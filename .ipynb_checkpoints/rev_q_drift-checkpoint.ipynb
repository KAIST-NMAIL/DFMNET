```python
%matplotlib inline
import numpy
from scipy.io import loadmat
from sklearn.metrics import mean_squared_error as MSE
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def getRMSE(tag):
    base_path = os.path.join(os.getcwd(), 'Result', tag+'_result.mat')
    dataset = loadmat(base_path)
    return np.sqrt(MSE(dataset['gt_raw'].T,dataset['pred_raw'].T, multioutput='raw_values'))
```

```python
def plots(rmse):
    plt.figure(figsize=(4,2), dpi=100)
    plt.plot(rmse)
    plt.xlabel('Frame')
    plt.ylabel('RMSE [m]')
```

```python
plots(getRMSE('squat'))
```
